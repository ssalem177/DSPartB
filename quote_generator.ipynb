```python
import pandas as pd
import numpy as np
import csv

# make code that runs each annotated file, turns it into a single speaker - addressee tibble

def book_read(book_name, prediction):

    if prediction == 'Correct':
    
        book = pd.read_csv('annotated_books/'+book_name+'annotated.csv')

    elif prediction == 'ML':

        book = pd.read_csv('predicted_books/ML_Full/'+book_name+'MLFullpredicted.csv')

    elif prediction == 'Baseline':

        book = pd.read_csv('predicted_books/baseline/'+book_name+'baselinepredicted.csv')

    # make addressees lists

    for i in book.index:

        if book.at[i,'dialogue'] != 'NO QUOTE':

            list_ad = []

            addressee = book.at[i,'addressee']

            if prediction != 'ML':

                for j in range(len(addressee)):
    
                    if addressee[j] == '\'':
    
                        word = ''
    
                        for k in range(j+1,len(addressee)):
    
                            if (addressee[k] != '\''):
    
                                word = word + addressee[k]
    
                            else:
                    
                                break
                    
            
                        list_ad.append(word)
    
    
                while ', ' in list_ad:
                        
                    list_ad.remove(', ')
    
                while ']' in list_ad:
                        
                    list_ad.remove(']')
                    
    
                book.at[i,'addressee'] = list_ad

    return book

Novel_names = []

with open('pdnc_dataset/PDNC-Novel-index.csv','r') as file:

    reader = csv.reader(file)

    for i in reader:

        novel_name = ""

        for j in i[1]:
            if ((j != " ") and (j != "'") and (j != "-")):
                novel_name = novel_name + j

        
        Novel_names.append(novel_name)
        
    Novel_names.remove('NovelTitle')
```

```python
# Making a modified version of the annotated books so that quotes that have more than one addressee are assigned as blanks

for i in Novel_names:

    
    
    book = book_read(i, prediction = 'Correct')
    
    new_database = []
    
    for j in book.index:
        
        if book.at[j,'dialogue'] != 'NO QUOTE':
        
            if len(book.at[j,'addressee']) == 1:
        
                new_row = book.loc[j]
        
                new_row.at['addressee'] = new_row.at['addressee'][0]
        
                new_database.append(new_row)

            else:

                new_row = book.loc[j]
        
                new_row.at['addressee'] = np.nan

                new_database.append(new_row)
                                  
        else:

            new_database.append(book.loc[j])

                                    
    
    
    new_database = pd.DataFrame(new_database)
    
    new_database.to_csv('Annotated_New/'+i+'anotatednew.csv')
    

for i in Novel_names:

    
    book = book_read(i, prediction = 'Baseline')

    book_correct = book_read(i, prediction = 'Correct')
    
    new_database = []

    # The predictions for baseline should correspond to single addressees as shown in the correct version
    
    for j in book.index:
        
        if book.at[j,'dialogue'] != 'NO QUOTE':
        
            if len(book_correct.at[j,'addressee']) == 1:

                if (len(book.at[j,'addressee']) == 1):
        
                    new_row = book.loc[j]
            
                    new_row.at['addressee'] = new_row.at['addressee'][0]
            
                    new_database.append(new_row)

                else:

                    new_row = book.loc[j]
            
                    new_row.at['addressee'] = np.nan
            
                    new_database.append(new_row)                    

            else:

                new_row = book.loc[j]
        
                new_row.at['addressee'] = np.nan

                new_database.append(new_row)
                                  
        else:

            new_database.append(book.loc[j])

                                    
    
    
    new_database = pd.DataFrame(new_database)
    
    new_database.to_csv('Baseline_new/'+i+'baselinenew.csv')
```

```python
for i in Novel_names:

    for type in ['ML','Correct','Baseline']:
        
        book = book_read(i, prediction = type)

        book_correct = book_read(i, prediction = 'Correct')
    
        new_database = []

        if type != 'ML':
    
            for j in book.index:
        
                if book.at[j,'dialogue'] != 'NO QUOTE':
        
                    if len(book_correct.at[j,'addressee']) == 1:
        
                        new_row = book.loc[j]
                        
                        if len(book.at[j,'addressee']) == 1:
        
                            new_row.at['addressee'] = new_row.at['addressee'][0]
        
                            new_database.append(new_row)

                        else:

                            new_database.append(new_row)

                        

        else:

            for j in book.index:

                if book.at[j,'dialogue'] != 'NO QUOTE':

                    if (book.at[j,'addressee'] is np.nan) == False:
    
                        new_row = book.loc[j]
    
                        new_database.append(new_row)
                
    
    
        new_database = pd.DataFrame(new_database)
    
        new_database.to_csv('Quotes/'+type+'/'+i+'quotes.csv')
        
```
