```python
from consolidation_ss import consolidation

import csv


# Getting all the novel names, but fixing names first.  

Novel_names = []

with open('pdnc_dataset/PDNC-Novel-index.csv','r') as file:

    reader = csv.reader(file)

    for i in reader:

        novel_name = ""

        for j in i[1]:
            
            if ((j != " ") and (j != "'") and (j != "-")):
                novel_name = novel_name + j

        
        Novel_names.append(novel_name)
        
    Novel_names.remove('NovelTitle')


# running annotations on all novels

for i in Novel_names:

    consolidation(i)
```
